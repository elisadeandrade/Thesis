### Response Distribution Graph

```python
def combine_dataframes(dfs):
    combined_df = pd.concat(dfs, axis=1)
    return combined_df

def plot_combined_all(df_neg_list, df_neu_list, df_pos_list, ax):
    df_negative = combine_dataframes(df_neg_list)
    df_neutral = combine_dataframes(df_neu_list)
    df_positive = combine_dataframes(df_pos_list)
    
    categories = pd.Index(df_negative.columns.str.replace('-adv', '').unique()).drop_duplicates()
    x_base = np.arange(len(categories)) * 45  # Increase spacing between categories

    offsets = {'Negative': -8, 'Neutral': 0, 'Positive': 8}
    line_styles = {'Negative': 'dashed', 'Neutral': 'dotted', 'Positive': 'solid'}
    dataframes = {'Negative': df_negative, 'Neutral': df_neutral, 'Positive': df_positive}

    ax.grid(which='both', color='lightgrey', linestyle='-', linewidth=0.5, zorder=0)

    for label, df in dataframes.items():
        offset = offsets[label]
        line_style = line_styles[label]

        ones_count = df.apply(lambda x: (x == 1).sum())
        
        normal_counts = ones_count[~ones_count.index.str.contains('-adv')]
        adv_counts = ones_count[ones_count.index.str.contains('-adv')]
        adv_counts.index = adv_counts.index.str.replace('-adv', '')

        # Remove duplicates from normal and adv counts
        normal_counts = normal_counts.groupby(normal_counts.index).sum()
        adv_counts = adv_counts.groupby(adv_counts.index).sum()

        # Align the counts with categories
        normal_counts = normal_counts.reindex(categories, fill_value=0)
        adv_counts = adv_counts.reindex(categories, fill_value=0)
        
        # Make 'Adv' counts negative
        adv_counts *= -1

        # Combine the counts
        combined_counts = pd.DataFrame({'Normal': normal_counts, 'Adv': adv_counts})

        x = x_base + offset 

        for idx, cat in enumerate(categories):
            # Normal points
            if combined_counts['Normal'][cat] != 0:
                ax.scatter([x[idx]], [combined_counts['Normal'][cat]], color='black', s=40, zorder=3)
                ax.vlines([x[idx]], 0, combined_counts['Normal'][cat], colors='black', linestyles=line_style, linewidth=2, zorder=4)

            # Adv points
            if combined_counts['Adv'][cat] != 0:
                ax.scatter([x[idx]], [combined_counts['Adv'][cat]], color='black', s=40, zorder=3)
                ax.vlines([x[idx]], 0, combined_counts['Adv'][cat], colors='black', linestyles=line_style, linewidth=2, zorder=4)

    ax.set_xticks(x_base)
    ax.set_xticklabels(categories, rotation=90, fontsize=10)
    ax.set_xlabel('Categories', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.set_title('Text 1 Response Distribution: Negative, Neutral, and Positive Prompts', fontsize=14)
    ax.axhline(0, color='black', linewidth=1, zorder=2)
    ax.set_ylim(-10, 16)
    ax.set_yticks(range(-10, 17, 2))

    handles = [plt.Line2D([0], [0], color='black', linestyle=line_styles[label], lw=4) for label in dataframes.keys()]
    labels = [f"{label} Prompt" for label in dataframes.keys()]
    ax.legend(handles, labels, fontsize=12)

df_neg_list = [df3]
df_neu_list = [df1]
df_pos_list = [df2]

fig, ax = plt.subplots(figsize=(15, 10), tight_layout=True)  
plot_combined_all(df_neg_list, df_neu_list, df_pos_list, ax)
plt.savefig('English_text1_NNP.png')
plt.show()
```

### Observations

**Normal Counts:**
- Categories like EBA (7), B (4), and IN (3) have higher counts with neutral prompts than negative prompts.
- Positive prompts have higher counts in categories like EBA (7), DS (4), IN (4), and INV (3) compared to negative prompts.
- Negative prompts have higher counts in categories like PP (13), CIT (4), CC (4), and HC (4) compared to positive and neutral prompts.

**Advice Counts:**
- Higher negative counts are observed in categories like RA-adv (2).
- Positive advice counts are higher in categories like SD-adv (4), EP-adv (2), and SE-adv (1).

**Key Observations:**
- The category PP shows a significant negative prompt count of 13.
- The category EBA has a consistent high count across both neutral and positive prompts (7 in both).
- There are more normal counts than advice counts overall.

### High-level category chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def combine_dataframes(dfs):
    combined_df = pd.concat(dfs, axis=1)
    return combined_df

def plot_combined_grouped(df_neg_list, df_neu_list, df_pos_list, ax):
    df_negative = combine_dataframes(df_neg_list)
    df_neutral = combine_dataframes(df_neu_list)
    df_positive = combine_dataframes(df_pos_list)
    
    # Define category groups
    category_groups = {
        'MJ': ['SE', 'SGE', 'SD', 'IN', 'CC', 'NC', 'HC', 'B'],
        'FA': ['EBA', 'FC', 'DS', 'INV'],
        'CA': ['PP', 'RA', 'EP', 'AP'],
        'P': ['CI', 'CIT', 'CY', 'CUW']
    }
    
    group_colors = {
        'MJ': '#3c78d8',
        'FA': '#9f6dec',
        'CA': '#f58617',
        'P': '#58a038'
    }
    
    offsets = {'Negative': -8, 'Neutral': 0, 'Positive': 8}
    line_styles = {'Negative': 'dashed', 'Neutral': 'dotted', 'Positive': 'solid'}
    dataframes = {'Negative': df_negative, 'Neutral': df_neutral, 'Positive': df_positive}

    ax.grid(which='both', color='lightgrey', linestyle='-', linewidth=0.5, zorder=0)
    
    for label, df in dataframes.items():
        offset = offsets[label]
        line_style = line_styles[label]

        ones_count = df.apply(lambda x: (x == 1).sum())
        
        normal_counts = ones_count[~ones_count.index.str.contains('-adv')]
        adv_counts = ones_count[ones_count.index.str.contains('-adv')]
        adv_counts.index = adv_counts.index.str.replace('-adv', '')

        normal_counts = normal_counts.groupby(normal_counts.index).sum()
        adv_counts = adv_counts.groupby(adv_counts.index).sum()

        # Group counts by category groups
        grouped_normal_counts = {group: normal_counts[normal_counts.index.isin(categories)].sum() for group, categories in category_groups.items()}
        grouped_adv_counts = {group: adv_counts[adv_counts.index.isin(categories)].sum() for group, categories in category_groups.items()}

        # Make 'Adv' counts negative
        for group in grouped_adv_counts:
            grouped_adv_counts[group] *= -1

        x_base = np.arange(len(category_groups)) * 45  # Increase spacing between categories
        combined_counts = pd.DataFrame({'Normal': grouped_normal_counts, 'Adv': grouped_adv_counts})

        x = x_base + offset 

        for idx, group in enumerate(category_groups):
            # Normal points
            if combined_counts['Normal'][group] != 0:
                ax.scatter([x[idx]], [combined_counts['Normal'][group]], color=group_colors[group], s=40, zorder=3)
                ax.vlines([x[idx]], 0, combined_counts['Normal'][group], colors=group_colors[group], linestyles=line_style, linewidth=2, zorder=4)

            # Adv points
            if combined_counts['Adv'][group] != 0:
                ax.scatter([x[idx]], [combined_counts['Adv'][group]], color=group_colors[group], s=40, zorder=3)
                ax.vlines([x[idx]], 0, combined_counts['Adv'][group], colors=group_colors[group], linestyles=line_style, linewidth=2, zorder=4)

    ax.set_xticks(x_base)
    ax.set_xticklabels(category_groups.keys(), rotation=90, fontsize=10)
    ax.set_xlabel('Categories', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.set_title('Detailed Analysis of English Prompts with High-Level Categories: Texts 1 to 5', fontsize=14)
    ax.axhline(0, color='black', linewidth=1, zorder=2)
    ax.set_ylim(-16, 16)
    ax.set_yticks(range(-50, 80, 5))

    category_handles = [plt.Line2D([0], [0], color=color, marker='o', linestyle='None', markersize=10) for color in group_colors.values()]
    category_labels = [f"{group}" for group in group_colors.keys()]
    
    style_handles = [plt.Line2D([0], [0], color='black', linestyle=line_styles[label], lw=4) for label in line_styles.keys()]
    style_labels = [f"{label} Prompt" for label in line_styles.keys()]

    handles = category_handles + style_handles
    labels = category_labels + style_labels
    
    ax.legend(handles, labels, fontsize=12)

    # Adding text annotations for Normal and Adv on the right side
    ax.text(1.05, 8, 'Normal', fontsize=12, color='black', verticalalignment='center', horizontalalignment='left', transform=ax.get_yaxis_transform())
    ax.text(1.05, -8, '-Adv', fontsize=12, color='black', verticalalignment='center', horizontalalignment='left', transform=ax.get_yaxis_transform())

df_neg_list = [df3, df2_3, df3_3, df4_3, df5_3]
df_neu_list = [df1, df2_1, df3_1, df4_1, df5_1]
df_pos_list = [df2, df2_2, df3_2, df4_2, df5_2]

fig, ax = plt.subplots(figsize=(15, 15), tight_layout=True)  
plot_combined_grouped(df_neg_list, df_neu_list, df_pos_list, ax)
plt.savefig('English_Higherlvl.png')
plt.show()
```
